```python
lista_encondings = [gabriel_encoding, matheus_encodings]
lista_nomes = ['Gabriel', 'Matheus']
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
imagem_teste = cv2.imread('/content/teste.jpeg')
image_rgb = cv2.cvtColor(imagem_teste, cv2.COLOR_BGR2RGB)
locazicacao_faces = face_recognition.face_locations(image_rgb, model='hog')
print(locazicacao_faces)
encoding_face = face_recognition.face_encodings(image_rgb, locazicacao_faces)
encoding_face = encoding_face[0]
print(encoding_face)
```

```python
resultados = face_recognition.compare_faces(lista_encondings, encoding_face)
print(resultados)
```

```python
distancias_faces = face_recognition.face_distance(lista_encondings, encoding_face)
print(distancias_faces)
```

```python
best_match_index = np.argmin(distancias_faces)
best_match_index
```

```python
if resultados[best_match_index]:
  pred = lista_nomes[best_match_index]
  print(pred + 'foi localizado na imagem')
  cv2_imshow(imagem_teste)
  print(distancias_faces[best_match_index])
```
